#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
result1 = employee.groupby(["Department", "Title"]).agg(Mean_Salary=('Salary', 'mean'),Num_Employees=('Name', 'count')).reset_index()
result1
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
result2 = employee.groupby(["Department", "Title"]).agg(Median_Salary=('Salary', 'median'),Min_Salary=('Salary', 'min'),Max_Salary=('Salary', 'max'),Std_Salary=('Salary', 'std')).reset_index()
result2
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here
result2['Std_Salary'] = result2['Std_Salary'].fillna(0)
result2
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
dpt_summ = employee.groupby("Department").agg(Num_Employees=('Name', 'count'),Total_Salary=('Salary', 'sum')).reset_index()
dpt_summ
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
output = employee.groupby(["Department", "Title"]).agg(Total_Years=('Years', 'sum'),Mean_Salary=('Salary', 'mean')).reset_index()
output
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here 

def mean_no_max(salary):
    return round(salary[salary != salary.max()].mean(),2)

(employee.pivot_table(values=['Salary'], index=['Department'], aggfunc={'Salary':mean_no_max}).rename(columns={'Salary': 'Mean salary without max'}))

```
